Сейчас тебе предстоит поучаствовать в викторине, которая поможет тебе подготовится к экзамену по Python. 
Всего в ней 35 вопросов по всем 10 темам, которые были в курсе.
В некоторых вопросах требуется выбрать один ответ, а в некоторых правильных ответов может быть несколько. Ответом на вопрос является цифра, 
соответствующая верному, на твой взгляд, варианту ответа, или последовательность цифр, если ответов несколько. 

При ответе на каждый вопрос у тебя будет 3 попытки.

Программа на протяжении прохождения викторины будет давать подсказки по вводу данных и дальнейшим действиям пользователя.

```python

# Авторы: Быкова Анна, Коровина Валерия, Полякова Дарья
import random
import sys

# создание класса, в котором будет храниться информация о каждом вопросе
class Question:
  def __init__(self):
    self.id = 0
    self.q_text = ''
    self.right_answ = [] 
    self.wrong_answ = []
    self.max_pts = 0

# создание класса, в котором будет храниться информация о каждом ответе
class Answer:
  def __init__(self):
    self.id = 0
    self.pts = 0
    self.given_answ = []

voprosy = {} # словарь для хранения всей информации о вопросе и вариантах ответа на него
otvety = {}  # словарь для хранения всей информации о ответах пользователя и полученных баллах
# формат хранения вопросов и ответов:
# текст вопроса; правильный ответ | правильный ответ...; неправильный ответ | неправильный ответ...
# создание функции для считывания информации о вопросах
def read_from_file(st):
  global voprosy, otvety
  f = open(st) # открытие файла с вопросами
  q_id = 0
  for question in f: # считывание информации о вопросе
    question = question.strip('\n')
    q = Question()   # создание объекта класса Question
    q.q_text = question.split('; ')[0] # записываем в переменную текст вопроса
    q.right_answ = question.split('; ')[1].split(' | ')  # записываем в переменную текст верных вариантов ответа
    q.wrong_answ = question.split('; ')[2].split(' | ') # записываем в переменную текст неверных вариантов ответа
    q.max_pts = len(q.right_answ) # записываем в переменную макс число баллов за вопрос по кол-ву верных ответов
    q.id = q_id # записываем в переменную id вопроса
    voprosy[q_id] = q # записываем в словарь по ключу id вопроса всю информацию о вопросе: текст, верные вар-ты, неверные вар-ты, макс число баллов
    q_id += 1
  f.close()
  return q_id


# создание функции для оценки и вывода итогового результата пользователя и его оценки (вывод и в консоль, и в файлик с отчетом)
def print_results(user_pts, maximum_pts, f):
  if user_pts/maximum_pts >= 0.8: 
    print('', file=f)
    print("На этом викторина окончена. Ты показал отличный результат!", file=f)
  elif 0.8 > user_pts/maximum_pts >= 0.6:
    print('', file=f)
    print("На этом викторина окончена. Ты показал хороший результат!", file=f)
  elif 0.6 > user_pts/maximum_pts >= 0.4:
    print('', file=f)
    print("На этом викторина окончена. Неплохой результат! Но ты можешь лучше:)", file=f)
  elif 0.4 > user_pts/maximum_pts >= 0:
    print('', file=f)
    print("На этом викторина окончена. Стоит повторить материал и пройти викторину заново:)", file=f)
  print(f"Доля правильных ответов: {round(user_pts/maximum_pts * 100, 2)}%", file=f)        
  print(f"Оценка: {round(user_pts / maximum_pts * 10)}", file=f)
  print('', file=f)


q_number = read_from_file('/content/gdrive/MyDrive/a.txt')
results = [] # создание списка для сохранения id ответов(у нас оно = id вопросов)


# ---User interaction-------------------------------------------
  
# вывод приветствия
print("""Привет! 
Сейчас тебе предстоит поучаствовать в викторине, 
которая поможет тебе подготовится к экзамену по Python. 
В некоторых вопросах требуется выбрать один ответ, 
а в некоторых правильных ответов может быть несколько. 
При ответе на каждый вопрос у тебя будет 3 попытки.
Чтобы начать, введи любой символ в следующее окно""")
print()
if input("Введи любой символ, чтобы начать тест: "):
  for id in range(q_number):
    q = voprosy[id]
    a = Answer() # создание объекта класса Answer
    a.id = q.id # присвоение ответу id вопроса для дальнейшего мэтчинга вопроса и ответа по id
    print(f'{id+1}. {q.q_text}') # вывод вопроса пользователю
    all_answers = q.right_answ + q.wrong_answ 
    random.shuffle(all_answers) # перемешивание вариантов ответа для вывода пользователю
    for i in range(len(all_answers)):
      print(f'  {i+1}. {all_answers[i]}') #вывод вариантов ответа пользователю
    attempts = 1  # создаем счетчик попыток
    while attempts <= 3: # устанавливаем лимит попыток = 3 для каждого вопроса
      given_answ = input('Введи все правильные варианты ответа через пробел: ')
      a.given_answ.append([]) # создание списка для сохранения текста ответов, данных пользователем
      met_right_answ = False
      for j in given_answ.split(): 
        if j.isdigit() and int(j) in range(1, len(all_answers) + 1): # проверяем, ввел ли пользователь корректные варианты ответа — цифры, входящие в номера вариантов ответа
          a.given_answ[-1].append(all_answers[int(j) - 1]) # сохранение текста ответов, данных пользователем, списком
          # подсчет баллов за вопросы
          if all_answers[int(j) - 1] in q.right_answ: 
            met_right_answ = True # обозначаем, встретился ли среди ответов пользователя верный
            a.pts += 1
          if all_answers[int(j) - 1] in q.wrong_answ:
            a.pts -= 1
        else:
          a.given_answ[-1].append('Введен некорректный вариант ответа') # сохранение в список некорректных ответов будет происходить в таком виде
      # выводим пользователю результат его ввода по набранному количеству очков
      if a.pts == q.max_pts:
        print('Твой ответ верный!')
        print()
        break
      elif met_right_answ == True and a.pts != q.max_pts: # рассматриваем случай, если есть неполное пересечение с набором верных ответов
        if attempts != 3: # вывод, если ответ частично верный и попытки еще есть
          print(f'Твой ответ частично верный. У тебя осталось {3 - attempts} попыток.')
        else: # вывод, если ответ частично верный, но попытки кончились
          print('Твой ответ частично верный. Попытки кончились.')
          print('Правильный ответ: ') #выводим пользователю правильный ответ на вопрос
          for i in q.right_answ:
            print('  ', i)
          print()
          print()
      else:
        if attempts != 3:  # вывод, если ответ неверный и попытки еще есть
          print(f'Ответ неверный. У тебя осталось {3 - attempts} попыток.')
        else:  # вывод, если ответ неверный, но попытки кончились
          print('Твой ответ неверный. Попытки кончились.')
          print('Правильный ответ: ') #выводим пользователю правильный ответ на вопрос
          for i in q.right_answ:
            print('  ', i)
          print()
      a.pts = 0
      attempts += 1
    otvety[a.id] = a # заполнение словаря со всей информацией об ответах пользователя
    results.append(a.id) # заполнение списка с информацией об id ответов (=id вопросов), чтобы обращаться затем по ним к словарю как по ключам

# ---Writing results-----------------------------------------------------------

user_pts = 0 # переменная для подсчета финальных баллов (=кол-ву сданных задач), набранных пользователем
maximum_pts = len(results) # переменная для подсчета максимума возможных набранных баллов
for i in results:
  if otvety[i].pts == voprosy[i].max_pts:
    user_pts += 1 # вытаскиваем из словаря информацию, сохраненную туда из объекта класса, по id ответа (=id вопроса)
  # maximum_pts += voprosy[i].max_pts # выстаскиваем из словаря информацию о максимуме возможных баллов на вопрос по id вопроса

print_results(user_pts, maximum_pts, sys.stdout) #выводим в консоль пользователю его результаты и потом благодарочку
print("Спасибо за участие и удачи на экзамене!")

#записываем в файл вопросы, заданные пользователю, все сделанные попытки, правильные ответы и кол-во баллов, набранных пользователем
with open ("/content/gdrive/MyDrive/answers_user.txt", "w", encoding = "utf-8") as f: 
  for id in results:
    print(f"Вопрос: {voprosy[id].q_text}", file = f)
    for j in range(len(otvety[id].given_answ)):
      print(f'Попытка №{j + 1}: ', file = f)
      for k in otvety[id].given_answ[j]:
        print(' ', k, file = f)
    print('Правильные ответы:', file = f)
    for h in voprosy[id].right_answ:
      print(' ', h, file = f)
    print('', file = f)
  print_results(user_pts, maximum_pts, f)


```
